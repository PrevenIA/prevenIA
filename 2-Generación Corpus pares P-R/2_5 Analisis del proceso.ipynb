Ahora vamos a analizar el rendimiento de los modelos.

```python
sf_mod = bronce1.groupby(['Modelo'])['Pregunta'].count()
```

```python
analisis2 = pd.DataFrame({'Modelo':sf_mod.index, 'Broncev1':sf_mod.values})
```

```python
analisis2
```

```python
sf2_mod = bronce2.groupby(['Modelo'])['Pregunta'].count()
aux = pd.DataFrame({'Modelo':sf2_mod.index, 'Broncev2':sf2_mod.values})
analisis2 = analisis2.join(aux.set_index('Modelo'), on='Modelo')
```

```python
analisis2
```

Calculamos ahora la diferencia:

```python
analisis2['Preguntas filtradas con Clasificación del dominio y similaridad semántica'] = analisis2['Broncev1'] - analisis2['Broncev2'] 
```

```python
analisis2
```

```python
sf3_mod = bronce3.groupby(['Modelo'])['Pregunta'].count()
aux = pd.DataFrame({'Modelo':sf3_mod.index, 'Broncev3':sf3_mod.values})
analisis2 = analisis2.join(aux.set_index('Modelo'), on='Modelo')
```

```python
analisis2
```

```python
analisis2['Preguntas filtradas manualmente por no experto'] = analisis2['Broncev2'] - analisis2['Broncev3'] 
analisis2['Filtrado total bronce desde crudo a definitivo'] = analisis2['Broncev1'] - analisis2['Broncev3'] 
```

```python
analisis2
```

```python
analisis2["Rendimiento del modelo en %"] = (1 - analisis2['Filtrado total bronce desde crudo a definitivo'] / analisis2["Broncev1"])*100
```

```python
analisis2
```

```python
grafico = analisis2[["Broncev1", "Broncev2", "Broncev3", "Modelo"]]
```

```python
grafico
```

```python
grafico = grafico.set_index('Modelo')
```

```python
grafico.plot(kind='barh')
```

```python
ax = grafico.plot(kind='barh')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
```
